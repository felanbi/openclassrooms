```python
SIZE_2D = (299, 299)
SIZE_3D = (299, 299, 3)
BATCH_SIZE = 32
```

```python
get_race_from_dir = lambda x: re.sub("n[0-9]+-", "", x).upper()
```

```python
val_dataset_full = tf.keras.utils.image_dataset_from_directory(
            '{}/Images_split/val'.format(root),
            image_size = SIZE_2D,
            batch_size=BATCH_SIZE,
            shuffle = False)
```

```python
model_transfer = load_model("{}/transfert_IV3.h5".format(root))
```

```python
def predict_image(img):

    img = img.reshape((-1,) + SIZE_3D) 
    pred = model_transfer.predict(img)[0]
    
    top5 = sorted(list(enumerate(pred)), reverse = True, key = lambda x: x[1])[:5]
    top5 = [(get_race_from_dir(val_dataset_full.class_names[x]), float(y)) for (x,y) in top5]

    return dict(top5)

image = gr.inputs.Image(shape = SIZE_2D)
label = gr.outputs.Label(num_top_classes=3)

gr.Interface(fn = predict_image, inputs = image, outputs=label,interpretation='default').launch(debug='True', share=True)
```
